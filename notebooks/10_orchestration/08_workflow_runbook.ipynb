# Pipeline — workflow runbook

## Create the Databricks Workflow

1. **Workflows** → **Create job** → import `config/workflows/globalmart_pipeline.job.json` or recreate tasks manually.
2. Replace `<REPO_PATH>` in each notebook path with your Repos path (e.g. `/Repos/you@email.com/ecommerce-analytics`).
3. Attach **Serverless** or a shared cluster to every task.
4. Pass job parameters: `pipeline_run_id`, `dry_run`, `simulate_failure`.

## Task graph

```text
bronze_ingestion
    → quality_checks
        → silver_transforms
            ├→ reconciliation ─┐
            └→ gold_aggregations ┴→ dimensional_refresh → visualization
```

## Successful run

Run with `simulate_failure` empty. Check `/Volumes/globalmart/metadata/run_reports/pipeline_*.json`.

## Failure demo (downstream skip)

Set job parameter **`simulate_failure`** = `silver_transforms` (or any task key). That task fails intentionally; **reconciliation**, **gold_aggregations**, **dimensional_refresh**, and **visualization** are skipped. Re-run with empty `simulate_failure` to recover.
